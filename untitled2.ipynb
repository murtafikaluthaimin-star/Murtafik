```python
# Jalankan ini di Google Colab (Runtime: Python3)
# Cell 1: install dependencies (opsional di Colab, biasanya sudah ada)
!pip install --quiet opencv-python-headless matplotlib numpy

# Cell 2: import & helper
import cv2
import numpy as np
import matplotlib.pyplot as plt
from google.colab import files
import os

plt.rcParams['figure.figsize'] = (12,8)

# buat folder output
os.makedirs('output', exist_ok=True)

# Cell 3: upload image (gunakan tombol file picker)
print("Upload satu citra grayscale (kakek.png). Jika citra berwarna, script akan konversi ke grayscale.")
uploaded = files.upload()
fn = list(uploaded.keys())[0]
print("File uploaded:", fn)

# baca citra
img_bgr = cv2.imdecode(np.frombuffer(uploaded[fn], np.uint8), cv2.IMREAD_UNCHANGED)
if img_bgr is None:
    raise ValueError("File tidak bisa dibaca. Pastikan file gambar valid.")
# Jika berwarna, konversi ke grayscale
if len(img_bgr.shape) == 3:
    img = cv2.cvtColor(img_bgr, cv2.COLOR_BGR2GRAY)
else:
    img = img_bgr.copy()

# pastikan dtype uint8
img = img.astype(np.uint8)

# Cell 4: transformasi negatif dan logaritmik
# Negatif
neg = 255 - img

# Logaritmik: G = c * log(1 + F), c = 255 / log(1 + max(F))
F = img.astype(np.float32)
maxF = float(F.max())
c = 255.0 / np.log(1.0 + maxF) if maxF > 0 else 1.0
log_img = c * np.log1p(F)   # log1p = log(1+F)
# normalisasi dan konversi ke uint8
log_img_u8 = np.clip(log_img, 0, 255).astype(np.uint8)

# simpan hasil
cv2.imwrite('output/original.png', img)
cv2.imwrite('output/negatif.png', neg)
cv2.imwrite('output/log.png', log_img_u8)

# Cell 5: tampilkan original, negatif, log dan histogram masing-masing (4 subplot grid)
fig, axes = plt.subplots(3, 4, figsize=(16,12))
# baris 1: gambar (original, negatif, log) plus placeholder
axes[0,0].imshow(img, cmap='gray'); axes[0,0].set_title('Original'); axes[0,0].axis('off')
axes[0,1].imshow(neg, cmap='gray'); axes[0,1].set_title('Negatif'); axes[0,1].axis('off')
axes[0,2].imshow(log_img_u8, cmap='gray'); axes[0,2].set_title('Logaritmik'); axes[0,2].axis('off')
axes[0,3].axis('off')

# baris 2: histogram masing-masing (bins 256)
bins = 256
axes[1,0].hist(img.ravel(), bins=bins); axes[1,0].set_title('Histogram - Original'); axes[1,0].set_xlim(0,255)
axes[1,1].hist(neg.ravel(), bins=bins); axes[1,1].set_title('Histogram - Negatif'); axes[1,1].set_xlim(0,255)
axes[1,2].hist(log_img_u8.ravel(), bins=bins); axes[1,2].set_title('Histogram - Logaritmik'); axes[1,2].set_xlim(0,255)
axes[1,3].axis('off')

# baris 3: tampilkan juga CDF (kumulatif) untuk membantu analisa
axes[2,0].plot(np.cumsum(np.histogram(img.ravel(), bins=bins, range=(0,255))[0])); axes[2,0].set_title('CDF - Original')
axes[2,1].plot(np.cumsum(np.histogram(neg.ravel(), bins=bins, range=(0,255))[0])); axes[2,1].set_title('CDF - Negatif')
axes[2,2].plot(np.cumsum(np.histogram(log_img_u8.ravel(), bins=bins, range=(0,255))[0])); axes[2,2].set_title('CDF - Logaritmik')
axes[2,3].axis('off')

plt.tight_layout()
plt.show()

# Cell 6: Soal 2  konvolusi dengan tiga kernel (LPF, HPF, BPF)
kern_lpf = (1.0/9.0) * np.array([[1,1,1],[1,1,1],[1,1,1]], dtype=np.float32)
kern_hpf = np.array([[-1,-1,-1],[-1,8,-1],[-1,-1,-1]], dtype=np.float32)
kern_bpf = np.array([[0,-1,0],[-1,5,-1],[0,-1,0]], dtype=np.float32)

# gunakan cv2.filter2D (preserve depth)
res_lpf = cv2.filter2D(img, ddepth=-1, kernel=kern_lpf, borderType=cv2.BORDER_REPLICATE)
res_hpf = cv2.filter2D(img, ddepth=-1, kernel=kern_hpf, borderType=cv2.BORDER_REPLICATE)
res_bpf = cv2.filter2D(img, ddepth=-1, kernel=kern_bpf, borderType=cv2.BORDER_REPLICATE)

# simpan hasil
cv2.imwrite('output/lpf.png', res_lpf)
cv2.imwrite('output/hpf.png', res_hpf)
cv2.imwrite('output/bpf.png', res_bpf)

# Cell 7: tampilkan original + 3 hasil filter dalam 1 figure
fig, ax = plt.subplots(1,4, figsize=(16,6))
ax[0].imshow(img, cmap='gray'); ax[0].set_title('Original'); ax[0].axis('off')
ax[1].imshow(res_lpf, cmap='gray'); ax[1].set_title('LPF (blur)'); ax[1].axis('off')
ax[2].imshow(res_hpf, cmap='gray'); ax[2].set_title('HPF (edge)'); ax[2].axis('off')
ax[3].imshow(res_bpf, cmap='gray'); ax[3].set_title('BPF (sharpen)'); ax[3].axis('off')
plt.tight_layout()
plt.show()

# hitung nilai rata-rata tiap gambar
means = {
    'original': float(np.mean(img)),
    'lpf': float(np.mean(res_lpf)),
    'hpf': float(np.mean(res_hpf)),
    'bpf': float(np.mean(res_bpf)),
    'negatif': float(np.mean(neg)),
    'logaritmik': float(np.mean(log_img_u8))
}
print("Rata-rata nilai piksel (mean):")
for k,v in means.items():
    print(f" - {k}: {v:.4f}")

# Cell 8: simpan analisa ke file .txt (isi analisa nanti bisa diganti)
analysis_text = f"""
ANALISA (otomatis disimpan oleh script)
File sumber: {fn}

1) Transformasi Nilai Keabuan:
 - c (untuk logaritmik) = {c:.6f}
 - Mean original: {means['original']:.4f}
 - Mean negatif: {means['negatif']:.4f}
 - Mean logaritmik: {means['logaritmik']:.4f}

2) Filter konvolusi:
 - Mean original: {means['original']:.4f}
 - Mean LPF (blur): {means['lpf']:.4f}
 - Mean HPF (edge): {means['hpf']:.4f}
 - Mean BPF (sharpen): {means['bpf']:.4f}

(Anda harus menambahkan analisa textual lebih lengkap sesuai soal. Lihat file analysis_manual.txt untuk analisa yang sudah saya siapkan.)
"""

with open('output/analysis_autogenerated.txt','w') as f:
    f.write(analysis_text)

# save a copy of manual analysis (detailed) which saya tulis di bawah (agar mudah di-upload)
manual_analysis = """=== ANALISA (MANUAL) ===

(Tempelkan seluruh bagian ANALISA yang saya kirim di sini; file ini juga sudah saya sediakan di akhir notebook.)
"""
with open('output/analysis_manual.txt','w') as f:
    f.write(manual_analysis)

print("Hasil tersimpan di folder 'output'. Anda bisa download file-file tersebut.")
# sediakan link download semua file output (Colab memungkinkan file zip)
!zip -r output.zip output >/dev/null
files.download('output.zip')
```

```python
# ============================
# Soal No. 2 - Konvolusi Citra (LPF, HPF, BPF)
# ============================

import cv2
import numpy as np
import matplotlib.pyplot as plt

# --- 1. Baca citra grayscale (gunakan citra yang sama dari Soal 1)
img = cv2.imread("kakek.png", cv2.IMREAD_GRAYSCALE)

# --- 2. Definisikan tiga kernel
lpf_kernel = (1/9) * np.array([[1,1,1],
                               [1,1,1],
                               [1,1,1]])

hpf_kernel = np.array([[-1,-1,-1],
                       [-1, 8,-1],
                       [-1,-1,-1]])

bpf_kernel = np.array([[ 0,-1, 0],
                       [-1, 5,-1],
                       [ 0,-1, 0]])

# --- 3. Terapkan konvolusi (filtering)
lpf_result = cv2.filter2D(img, -1, lpf_kernel)
hpf_result = cv2.filter2D(img, -1, hpf_kernel)
bpf_result = cv2.filter2D(img, -1, bpf_kernel)

# --- 4. Tampilkan hasil dalam satu figure
plt.figure(figsize=(12,8))

plt.subplot(2,2,1)
plt.imshow(img, cmap='gray')
plt.title("Original")
plt.axis('off')

plt.subplot(2,2,2)
plt.imshow(lpf_result, cmap='gray')
plt.title("Low Pass Filter (LPF)")
plt.axis('off')

plt.subplot(2,2,3)
plt.imshow(hpf_result, cmap='gray')
plt.title("High Pass Filter (HPF)")
plt.axis('off')

plt.subplot(2,2,4)
plt.imshow(bpf_result, cmap='gray')
plt.title("Band Pass Filter (BPF)")
plt.axis('off')

plt.tight_layout()
plt.show()

# --- 5. Hitung rata-rata nilai piksel
mean_original = np.mean(img)
mean_lpf = np.mean(lpf_result)
mean_hpf = np.mean(hpf_result)
mean_bpf = np.mean(bpf_result)

print("Rata-rata nilai piksel (mean):")
print(f" - Original: {mean_original:.4f}")
print(f" - LPF: {mean_lpf:.4f}")
print(f" - HPF: {mean_hpf:.4f}")
print(f" - BPF: {mean_bpf:.4f}")
```
