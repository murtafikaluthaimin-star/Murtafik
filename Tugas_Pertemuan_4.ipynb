```python
import cv2
import matplotlib.pyplot as plt

img = cv2.imread('/ccontoh 1.jpg', cv2.IMREAD_GRAYSCALE)
plt.imshow(img, cmap='gray')
plt.title('Citra Asli (Grayscale)')
plt.show()

# Citra negatif
negative = 255 - img
plt.imshow(negative, cmap='gray')
plt.title('Negative Image')
plt.show()

# 1. Citra biner (thresholding)
threshold_value = 128
_, binary = cv2.threshold(img, threshold_value, 255, cv2.THRESH_BINARY)
plt.imshow(binary, cmap='gray')
plt.title('Citra Biner (Thresholding)')
plt.show()

# 2. Mencerahkan gambar (menambah intensitas)
bright = cv2.add(img, 50)
plt.imshow(bright, cmap='gray')
plt.title('Citra Lebih Terang')
plt.show()

# 3. Menggelapkan gambar (mengurangi intensitas)
dark = cv2.subtract(img, 50)
plt.imshow(dark, cmap='gray')
plt.title('Citra Lebih Gelap')
plt.show()

# 4. Menampilkan histogram citra asli
plt.hist(img.ravel(), bins=256, range=[0, 256])
plt.title('Histogram Citra Asli')
plt.xlabel('Intensitas Pixel')
plt.ylabel('Jumlah Pixel')
plt.show()
```

```python
import cv2
import matplotlib.pyplot as plt

img = cv2.imread('/ccontoh 1.jpg', cv2.IMREAD_GRAYSCALE)
plt.imshow(img, cmap='gray')
plt.title('Citra Asli (Grayscale)')
plt.show()

# Citra negatif
negative = 255 - img
plt.imshow(negative, cmap='gray')
plt.title('Negative Image')
plt.show()

# 1. Citra biner (thresholding)
threshold_value = 128
_, binary = cv2.threshold(img, threshold_value, 255, cv2.THRESH_BINARY)
plt.imshow(binary, cmap='gray')
plt.title('Citra Biner (Thresholding)')
plt.show()

# 2. Mencerahkan gambar (menambah intensitas)
bright = cv2.add(img, 50)
plt.imshow(bright, cmap='gray')
plt.title('Citra Lebih Terang')
plt.show()

# 3. Menggelapkan gambar (mengurangi intensitas)
dark = cv2.subtract(img, 50)
plt.imshow(dark, cmap='gray')
plt.title('Citra Lebih Gelap')
plt.show()

# 4. Menampilkan histogram citra asli
plt.hist(img.ravel(), bins=256, range=[0, 256])
plt.title('Histogram Citra Asli')
plt.xlabel('Intensitas Pixel')
plt.ylabel('Jumlah Pixel')
plt.show()

# Membaca gambar dalam mode grayscale
img = cv2.imread('ccontoh 1.jpg', cv2.IMREAD_GRAYSCALE)

# 1. Citra negatif
negative = 255 - img

# 2. Citra biner (thresholding)
threshold_value = 128
_, binary = cv2.threshold(img, threshold_value, 255, cv2.THRESH_BINARY)

# 3. Citra lebih terang
bright = cv2.add(img, 50)

# 4. Citra lebih gelap
dark = cv2.subtract(img, 50)

# --- Gabungkan semua hasil dalam satu figure ---
plt.figure(figsize=(12, 8))

plt.subplot(2, 3, 1)
plt.imshow(img, cmap='gray')
plt.title('Citra Asli')
plt.axis('off')

plt.subplot(2, 3, 2)
plt.imshow(negative, cmap='gray')
plt.title('Citra Negatif')
plt.axis('off')

plt.subplot(2, 3, 3)
plt.imshow(binary, cmap='gray')
plt.title('Citra Biner')
plt.axis('off')

plt.subplot(2, 3, 4)
plt.imshow(bright, cmap='gray')
plt.title('Citra Terang')
plt.axis('off')

plt.subplot(2, 3, 5)
plt.imshow(dark, cmap='gray')
plt.title('Citra Gelap')
plt.axis('off')

plt.subplot(2, 3, 6)
plt.hist(img.ravel(), bins=256, range=[0, 256], color='black')
plt.title('Histogram Citra Asli')

plt.tight_layout()
plt.show()
```

```python
# --- LANJUTAN DARI KODE SEBELUMNYA ---

# 5. Menghaluskan citra (Smoothing / Blur)
blur = cv2.GaussianBlur(img, (5, 5), 0)
plt.imshow(blur, cmap='gray')
plt.title('Citra Setelah Gaussian Blur')
plt.axis('off')
plt.show()

# 6. Menajamkan citra (Sharpening)
import numpy as np
kernel_sharpen = np.array([[0, -1, 0],
                           [-1, 5, -1],
                           [0, -1, 0]])
sharpen = cv2.filter2D(img, -1, kernel_sharpen)
plt.imshow(sharpen, cmap='gray')
plt.title('Citra Setelah Sharpening')
plt.axis('off')
plt.show()

# 7. Deteksi tepi (Edge Detection)
edges = cv2.Canny(img, 100, 200)
plt.imshow(edges, cmap='gray')
plt.title('Citra Deteksi Tepi (Canny Edge)')
plt.axis('off')
plt.show()

# 8. Equalisasi Histogram (meningkatkan kontras)
equalized = cv2.equalizeHist(img)
plt.imshow(equalized, cmap='gray')
plt.title('Citra Setelah Equalisasi Histogram')
plt.axis('off')
plt.show()

# 9. Gabungkan semua hasil filtering dan enhancement
plt.figure(figsize=(12, 8))

plt.subplot(2, 3, 1)
plt.imshow(img, cmap='gray')
plt.title('Citra Asli')
plt.axis('off')

plt.subplot(2, 3, 2)
plt.imshow(blur, cmap='gray')
plt.title('Gaussian Blur')
plt.axis('off')

plt.subplot(2, 3, 3)
plt.imshow(sharpen, cmap='gray')
plt.title('Sharpening')
plt.axis('off')

plt.subplot(2, 3, 4)
plt.imshow(edges, cmap='gray')
plt.title('Edge Detection')
plt.axis('off')

plt.subplot(2, 3, 5)
plt.imshow(equalized, cmap='gray')
plt.title('Histogram Equalization')
plt.axis('off')

plt.tight_layout()
plt.show()

# 10. (Opsional) Simpan hasil ke file
cv2.imwrite('hasil_negative.jpg', negative)
cv2.imwrite('hasil_binary.jpg', binary)
cv2.imwrite('hasil_bright.jpg', bright)
cv2.imwrite('hasil_dark.jpg', dark)
cv2.imwrite('hasil_blur.jpg', blur)
cv2.imwrite('hasil_sharpen.jpg', sharpen)
cv2.imwrite('hasil_edges.jpg', edges)
cv2.imwrite('hasil_equalized.jpg', equalized)

print(" Semua hasil telah ditampilkan dan disimpan ke file.")
```
